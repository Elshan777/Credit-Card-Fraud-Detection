### There are 30 columns in dataset. Class is target column. It can either be 0 (non fraud) or 1 (fraud).

```python
from collections import Counter
```

### Our data is imbalanced

```python
Counter(df['Class'])
```

### There are 492 fraud and 284315 non fraud observations on the dataset.
### There are few things we could do 
- Oversample
- Undersample
- SMOTE

### If we do not balance our data our accuracy will be high even though we mark all fraud cases as non-fraud. And we don't want that

### The first thing we will try is to undersample the data and we'll use RandomUnderSample for this 

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
X = df.iloc[:, :-1]
y = df['Class']
```

```python
und = RandomUnderSampler()
X_und, y_und = und.fit_resample(X, y)
```

```python
Counter(y_und)
```

### We undersampled data and there are equal amounts of Fraud and Non fraud observations

```python
# Lets concat undersampled X and y
df_und = pd.concat([X_und, y_und], axis=1)
df_und.head()
```

```python
df_und.hist(figsize = (20, 20))
plt.show()
```

## In the next step we will visualize by reducing dimensions

### I will use methods such PCA, t-SNE and UMAP

```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from umap import UMAP
import time
import matplotlib.patches as mpatches
```

```python
df_und.head()
```

```python
X_und = df_und.iloc[:, :-1]
y_und = df_und.iloc[:, -1]
```

```python
# T-SNE Implementation
t0 = time.time()
X_reduced_tsne = TSNE(n_components=2, random_state=42).fit_transform(X_und.values)
t1 = time.time()
print("T-SNE took {:.2} s".format(t1 - t0))

# PCA Implementation
t0 = time.time()
X_reduced_pca = PCA(n_components=2, random_state=42).fit_transform(X_und.values)
t1 = time.time()
print("PCA took {:.2} s".format(t1 - t0))

# TruncatedSVD
t0 = time.time()
X_reduced_umap = UMAP(n_components=2).fit_transform(X_und.values)
t1 = time.time()
print("UMAP took {:.2} s".format(t1 - t0))
```

```python
f, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(24,6))
# labels = ['No Fraud', 'Fraud']
f.suptitle('Clusters using Dimensionality Reduction', fontsize=14)


blue_patch = mpatches.Patch(color='#0A0AFF', label='0')
red_patch = mpatches.Patch(color='#AF0000', label='1')


# t-SNE scatter plot
ax1.scatter(X_reduced_tsne[:,0], X_reduced_tsne[:,1], c=(y_und == 0), cmap='coolwarm', label='0', linewidths=2)
ax1.scatter(X_reduced_tsne[:,0], X_reduced_tsne[:,1], c=(y_und == 1), cmap='coolwarm', label='1', linewidths=2)
ax1.set_title('t-SNE', fontsize=14)

ax1.grid(True)

ax1.legend(handles=[blue_patch, red_patch])


# PCA scatter plot
ax2.scatter(X_reduced_pca[:,0], X_reduced_pca[:,1], c=(y_und == 0), cmap='coolwarm', label='0', linewidths=2)
ax2.scatter(X_reduced_pca[:,0], X_reduced_pca[:,1], c=(y_und == 1), cmap='coolwarm', label='1', linewidths=2)
ax2.set_title('PCA', fontsize=14)
ax2.grid(True)

ax2.legend(handles=[blue_patch, red_patch])

# TruncatedSVD scatter plot
ax3.scatter(X_reduced_umap[:,0], X_reduced_umap[:,1], c=(y_und == 0), cmap='coolwarm', label='0', linewidths=2)
ax3.scatter(X_reduced_umap[:,0], X_reduced_umap[:,1], c=(y_und == 1), cmap='coolwarm', label='1', linewidths=2)
ax3.set_title('UMAP', fontsize=14)

ax3.grid(True)

ax3.legend(handles=[blue_patch, red_patch])

plt.show()
```

```python
from sklearn.decomposition import KernelPCA
from sklearn.manifold import Isomap
```

```python
t0 = time.time()
X_reduced_pca_rbf = KernelPCA(n_components = 2, kernel="rbf").fit_transform(X_und.values)
t1 = time.time()
print("RBF took {:.2} s".format(t1 - t0))

t0 = time.time()
X_reduced_pca_sigmoid = KernelPCA(n_components = 2, kernel="sigmoid",).fit_transform(X_und.values)
t1 = time.time()
print("Sigmoid took {:.2} s".format(t1 - t0))

t0 = time.time()
X_reduced_isomap = Isomap(n_components=2).fit_transform(X_und.values)
t1 = time.time()
print("Isomap took {:.2} s".format(t1 - t0))
```

```python
fi, (ax4, ax5, ax6) = plt.subplots(1, 3, figsize=(24,6))
fi.suptitle('Clusters using Dimensionality Reduction', fontsize=14)


ax4.scatter(X_reduced_pca_rbf[:,0], X_reduced_pca_rbf[:,1], c=(y_und == 0), cmap='coolwarm', label='0', linewidths=2)
ax4.scatter(X_reduced_pca_rbf[:,0], X_reduced_pca_rbf[:,1], c=(y_und == 1), cmap='coolwarm', label='1', linewidths=2)
ax4.set_title('PCA RBF', fontsize=14)

ax4.grid(True)

ax4.legend(handles=[blue_patch, red_patch])


ax5.scatter(X_reduced_pca_sigmoid[:,0], X_reduced_pca_sigmoid[:,1], c=(y_und == 0), cmap='coolwarm', label='0', linewidths=2)
ax5.scatter(X_reduced_pca_sigmoid[:,0], X_reduced_pca_sigmoid[:,1], c=(y_und == 1), cmap='coolwarm', label='1', linewidths=2)
ax5.set_title('PCA Sigmoid', fontsize=14)

ax5.grid(True)

ax5.legend(handles=[blue_patch, red_patch])

# TruncatedSVD scatter plot
ax6.scatter(X_reduced_isomap[:,0], X_reduced_isomap[:,1], c=(y_und == 0), cmap='coolwarm', label='0', linewidths=2)
ax6.scatter(X_reduced_isomap[:,0], X_reduced_isomap[:,1], c=(y_und == 1), cmap='coolwarm', label='1', linewidths=2)
ax6.set_title('Isomap', fontsize=14)

ax6.grid(True)

ax6.legend(handles=[blue_patch, red_patch])

plt.show()
```
